```python
# 1. What is Simple Linear Regression (SLR)? Explain its purpose.
'''
Simple Linear Regression (SLR) is a statistical and machine learning technique used to model the linear relationship 
between one independent variable (X) and one dependent variable (Y) by fitting a straight line to the observed data.
Purpose of Simple Linear Regression -
1. Helps identify how one variable affects another
2. Used to predict the value of Y (dependent variable) for a given X (independent variable)
3. Measures how much Y changes when X increases by one unit
4. Detects increasing or decreasing trends in data
5. Supports data-driven business decisions
'''
```

```python
# 2. What are the key assumptions of Simple Linear Regression?
'''
1. There must be a linear relationship between the independent variable (X) and the dependent variable (Y) .
2. The residuals (errors) must be independent of each other .
3. The variance of residuals should be constant across all values of X.
4. The residuals should be normally distributed.
5. In SLR, there is only one predictor, so this assumption is automatically satisfied.
'''
```

```python
# 3. Write the mathematical equation for a simple linear regression model and explain each term.
'''
Y = B0 + B1X + E
Y = Dependent Variable
X = Independent Variable
B0 = Intercept
B1 = Slope
E = Error term
'''
```

```python
# 4. Provide a real-world example where simple linear regression can be applied.
'''
Predicting House Price Based on Area - 
Simple linear regression can be used to predict the price of a house based on its area in square feet.
Independent variable (X): Area of the house
Dependent variable (Y): House price
Price = B0 + B1 * Area
This helps estimate the price of a new house given its area.
It assists real estate companies in pricing strategies.
It supports buyers and sellers in making data-driven decisions.
B1 shows how much the house price increases for every additional square foot.
B0 represents the base price when area is zero.
'''
```

```python
# 5. What is the method of least squares in linear regression?
'''
The method of least squares is a mathematical technique used in linear regression to estimate 
the best fitting regression line by minimizing the sum of the squared differences 
between the actual values and the predicted values.
It finds the line that comes closest to all data points.
It produces the best linear unbiased estimator under standard regression assumptions.
'''
```

```python
# 6. What is Logistic Regression? How does it differ from Linear Regression?
'''
Logistic Regression is a supervised machine learning algorithm used for binary classification problems 
where the target variable has only two possible outcomes (e.g., 0 or 1, Yes or No, Pass or Fail).
Logistic regression predicts the probability that an observation belongs to a particular class.
1. Linear Regression predict continuous values.
   Logistic Regression predict classes.
2. Linear Regression output is continuous.
   Logistic Regression output is 0 or 1 (Probability) .
3. The problem type for linear is regression but for logistic is classification.
4. Error distribution in linear is normally distributed but in logistic is binomial distribution.
'''
```

```python
# 7. Name and briefly describe three common evaluation metrics for regression models.
'''
1. Mean Absolute Error - 
   Measures the average absolute difference between actual and predicted values.
   Easy to interpret because it is in the same units as the target variable.
2. Mean Squared Error -
   Measures the average of squared errors between actual and predicted values.
   Penalizes large errors more heavily.
3. Root Mean Squared Error - 
   Interpretable in the same units as the target variable.
Common regression evaluation metrics include - 
MAE measures average absolute error 
MSE penalizes larger errors by squaring them
RMSE provides error magnitude in the original units of the target variable.
'''
```

```python
# 8. What is the purpose of the R-squared metric in regression analysis?
'''
R-squared is a statistical measure that indicates how well a regression model explains 
the variability of the dependent variable using the independent variables.
Purpose of R-squared - 
1. Measure Goodness of Fit -
   Shows the proportion of variance in the dependent variable explained by the model.
2. Model Comparison -
   Helps compare multiple regression models.
3. Explain Predictive Power -
   Indicates how useful the predictors are in explaining the outcome.
4. Baseline Model Evaluation -
   Tells whether the model performs better than simply predicting the mean.
The purpose of the R-squared metric is to measure how much of the variation in the dependent variable 
is explained by the regression model, thereby indicating the goodness of fit.
'''
```

```python
# 9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 6])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)
```

```python
# 10. How do you interpret the coefficients in a simple linear regression model.
'''
Y = B0 + B1X + E
B0 (Intercept) - 
   Represents the expected value of Y when X = 0.
   It is the point where the regression line crosses the Y-axis.
B1 (Slope) -
   Represents the average change in Y for a one-unit increase in X.
   Indicates the direction and strength of the relationship.
In a simple linear regression model, 
the intercept represents the expected value of the dependent variable when the independent variable is zero, 
while the slope represents the change in the dependent variable for a one-unit change in the independent variable.
'''
```
